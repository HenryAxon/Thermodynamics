```python
import numpy as np
import matplotlib as plot
%matplotlib inline
import random
from matplotlib import pyplot as plt



def randomwalk(steps):
    x = 0
    start = x
    position = [x]
    probs = [1,-1]
    for i in range(1, steps):
        x += random.choice(probs)
        position.append(x)
    return position

def multiplewalks(walkers):
    numberofwalkers = walkers
    path = []
    steps = 100000
    for walker in range(walkers):
        walker_path=[0]
        position = 0
        for i in range(0, steps):
            position += random.choice([-1,1])
            walker_path.append(position)
        path.append(walker_path)
    path = np.array(path)
    meansd = [np.mean(path[:, i]**2) for i in range(1, steps)]
    return meansd

walk1 = np.array(randomwalk(100000))
plt.figure(figsize = (10,10))
plt.title("Random Walk 1")
plt.xlabel("Step")
plt.ylabel('Position')
plt.plot(walk1)

displacementsquare1 = (walk1 ** 2)

plt.figure(figsize = (10,10))
plt.xlabel("Steps")
plt.ylabel('Displacement^2')
plt.title('Displacement Squared vs Time ')
plt.plot(displacementsquare1)


#Below is the code for the mean distance squared plots 
walkers = 10
path = []
steps = 100000
for walker in range(walkers):
    walker_path=[0]
    position = 0
    for i in range(0, steps):
        position += random.choice([-1,1])
        walker_path.append(position)
    path.append(walker_path)
path = np.array(path)
meansd = [np.mean(path[:, i]**2) for i in range(0, steps)]        
    
plt.figure(figsize = (10,10))
plt.xlabel("Steps")
plt.ylabel('Mean Displacement^2')
plt.title('Mean Displacement Squared vs Time for 10 Walkers ')
plt.plot(meansd)

walkers = 100
path = []
for walker in range(walkers):
    walker_path=[0]
    position = 0
    for i in range(0, steps):
        position += random.choice([-1,1])
        walker_path.append(position)
    path.append(walker_path)
path = np.array(path)
meansd = [np.mean(path[:, i]**2) for i in range(0, steps)]

plt.figure(figsize=(10,10))
plt.xlabel("Steps")
plt.ylabel('Mean Displacement^2')
plt.title('Mean Displacement Squared vs Time for 100 Walkers ')
plt.plot(meansd)

walkers = 1000
path = []
steps = 100000
for walker in range(walkers):
    walker_path=[0]
    position = 0
    for i in range(0, steps):
        position += random.choice([-1,1])
        walker_path.append(position)
    path.append(walker_path)
path = np.array(path)
meansd = [np.mean(path[:, i]**2) for i in range(0, steps)]


plt.figure(figsize = (10,10))
plt.xlabel("Steps")
plt.ylabel('Mean Displacement^2')
plt.title('Mean Displacement Squared vs Time for 1000 Walkers ')
plt.plot(meansd)
```
